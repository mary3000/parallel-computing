```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
def get_data(file):
    time_data = []
    with open(file, newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
        for row in spamreader:
            time_data.append(row)
    procs = list((int(i[0]) for i in time_data))
    times = list((float(i[1]) for i in time_data))
    print(time_data[0][2])
    procs.reverse()
    times.reverse()
    times = np.array(times)
    procs = np.array(procs)
    return procs, times
```

```python
def draw_speedup(file):
    plt.figure(figsize=(16, 6))
    procs, times = get_data(file)
    plt.plot(procs, procs, label='y = x')
    plt.xlabel('Number of processes')
    plt.ylabel('Speedup, time(1) / time(x)')
    plt.scatter(procs, times[0] / times, color='red', label='Speedup')
    plt.legend()
    plt.show()
```

```python
draw_speedup('time_cs.csv')
```

```python
draw_speedup('time_red.csv')
```
