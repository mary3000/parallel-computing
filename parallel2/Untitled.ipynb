```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
sample = []
with open('res_new1.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        sample.append(row)
```

```python
points = []
with open('points.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        points.append(row)
x = list((float(i[0]) for i in points))
y = list((int(i[1]) for i in points))
```

```python
for i in range(len(sample)):
    for j in range(len(sample[0])):
        sample[i][j] = float(sample[i][j])
```

```python
grid = np.arange(0, 0.5, step=0.005)
plt.figure(figsize=(16, 6))
plt.ylim((0, 100))
plt.xlim((0, 0.5))
plt.xlabel('Distance, meters')
plt.ylabel('Temperature, C')
#plt.plot(grid, sample[len(sample) // 2])

a = int(160 // 5)
b = int(320 // 5)
arr_sample = np.array(sample)
plt.plot(grid, sample[a], label='x=0.16m')
plt.plot(grid, sample[b], label='x=0.32m')
plt.plot(grid, arr_sample[:, a], label='y=0.16m')
plt.plot(grid, arr_sample[:, b], label='y=0.32m')
plt.legend()

#plt.scatter(x, y, color='red')
plt.show()
```

```python
def draw3D(x, y, z):
    fig = plt.figure(figsize=(12, 12))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap=cm.Spectral_r)
    ax.set_xlabel('X, meters')
    ax.set_ylabel('Y, meters')
    ax.set_zlabel('Temperature, C')
    plt.show()
```

```python
sample = np.array(sample)
X, Y = np.meshgrid(grid, grid)
draw3D(X, Y, sample)
```

```python
time_data = []
with open('time_new1.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        time_data.append(row)
procs = list((int(i[0]) for i in time_data))
times = list((float(i[1]) for i in time_data))
```

```python
time_data1 = []
with open('time.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        time_data1.append(row)
procs1 = list((int(i[0]) for i in time_data1))
times1 = list((float(i[1]) for i in time_data1))
```

```python
plt.figure(figsize=(16, 6))
times = np.array(times)
procs = np.array(procs)
times1 = np.array(times1)
procs1 = np.array(procs1)
procs = procs - 1
plt.xlabel('Number of processes')
plt.ylabel('Speedup, time(1) / time(x)')
plt.plot(procs, procs, label='y = x')
plt.scatter(procs, times[0] / times, color='red', label='Speedup')
plt.scatter(procs1, times1[0] / times1, color='green', label='Speedup')
plt.legend()
plt.show()
```
